```python
import numpy as np
```

# ARRAY

#### 1D ARRAY

```python
array_1d= np.array([1,3,5,7,9])
array_1d
```

#### 2D ARRAY

```python
array_2d=np.array([[1,2,3],[2,3,3],[3,3,3]])
print(array_2d)
```

#### 3D ARRAY

```python
array_3d=np.array([[[1,2,3,6],
                   [2,3,3,7],
                   [3,3,3,8]],
                   [[3,3,3,6],
                    [5,6,7,5],
                    [5,5,5,7]],
                    [[3,3,3,4],
                    [5,6,7,5],
                    [5,5,5,5]]])
print(array_3d)
```
